```python
import joblib

spam_pack = joblib.load('../Model/spam_file_scanner.pkl')
print("Keys in file Spam:", spam_pack.keys())

malware_pack = joblib.load('../Model/malware_rf_model.pkl')
print("Keys in file Malware:", malware_pack.keys())

news_pack = joblib.load('../Model/ultimate_news_detector.pkl')
print("Keys in file Newspaper:", news_pack.keys())
```

```python
# 1. Tách cho Spam (Dùng 'cv' đúng như trong file của ông)
rf_spam = spam_pack['model']
vec_spam = spam_pack['cv'] 

# 2. Tách cho Malware
rf_malware = malware_pack['model']
vec_malware = malware_pack['vectorizer']

# 3. Tách cho Newspaper
rf_news = news_pack['model']
vec_news = news_pack['vectorizer']

print("✅ Đã tách xong! Biến 'vec_spam' hiện tại đã sẵn sàng.")
```

```python
import re

# --- 1. Hàm dự đoán Spam ---
# --- Hàm dự đoán Spam (Đã sửa lỗi) ---
def predict_spam(text):
    # 1. Chuyển text thành vector bằng 'cv' (CountVectorizer)
    # Nhớ bọc [text.lower()] để nó không bị lỗi "Iterable expected"
    X = vec_spam.transform([text.lower()])
    
    # 2. Dự đoán
    pred = rf_spam.predict(X)[0]
    
    # 3. Ép kiểu về String rồi mới dùng .upper() để không bị lỗi 'int' object
    # Đồng thời check nếu là số 0/1 thì map về nhãn Ham/Spam cho đẹp
    result = str(pred).lower()
    if result == '1' or result == 'spam':
        return "Spam Check: SPAM 🛑"
    else:
        return "Spam Check: HAM (SAFE) ✅"

# --- 2. Hàm dự đoán Malware ---
def predict_malware(code):
    X = vec_malware.transform([code.lower()])
    pred = rf_malware.predict(X)[0]
    status = "MALICIOUS 🛑" if str(pred) == '1' else "SAFE ✅"
    return f"Malware Check: {status}"

# --- 3. Hàm dự đoán Newspaper ---
def predict_news(url):
    url_clean = re.sub(r'https?://|www\.', '', url.lower()).strip()
    X = vec_news.transform([url_clean])
    pred = rf_news.predict(X)[0]
    status = "REAL NEWS ✅" if pred in ['Real', 'Safe'] else "FAKE/PHISHING LINK 🛑"
    return f"Link Check: {status}"

print("🚀 Toàn bộ hàm đã cập nhật chuẩn 100%!")
```

```python
print(predict_spam("Ccongratulations! you've won a free iPhone. click here to claim now!"))
print(predict_malware("import os; os.system('rm -rf /')"))
print(predict_news("https://vnexpress.net/thoi-su"))
```
