```python
import pandas as pd

# Read the CSV file from the Processed folder
df = pd.read_csv('../Processed/final_training_data.csv')

# Split the dataframe into 5 parts (adjust n as needed)
n = 5
chunk_size = len(df) // n

for i in range(n):
    start = i * chunk_size
    end = (i + 1) * chunk_size if i < n - 1 else len(df)
    chunk = df.iloc[start:end]
    chunk.to_csv(f'Dataset0{i+5}_cleaned.csv', index=False)
```
